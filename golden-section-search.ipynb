```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def golden_section_search(f, xl0, xr0, true_err_max = 0.0, i_max = 100, debug = False, plot = False):
    """
    Finds a local maximum of a function on a bounded interval.
    """
    
    # golden ratio
    phi = (np.sqrt(5.0) + 1.0) / 2.0
    
    # initial values
    xll = min(xl0, xr0)
    xrr = max(xl0, xr0)
    i = 0
    true_err = xrr - xll
    
    # track values
    xls = []
    xrs = []
    
    while true_err > true_err_max and i < i_max:
        # iteration number
        i += 1
        
        # compute interior points
        xl = xrr - (xrr - xll) / phi
        xr = xll + (xrr - xll) / phi
        fxl = f(xl)
        fxr = f(xr)
        
        # track values
        xls.append(xll)
        xrs.append(xrr)

        # update
        if fxl > fxr:
            xrr = xr
        else:
            xll = xl
        
        # compute error measures
        true_err = xrr - xll
        
        # debug if specified
        if debug:
            print('--------------------')
            print(f'Iteration: {i}')
            print(f'Left: {xll:.6g}')
            print(f'Right: {xrr:.6g}')
    
    # plot if specified
    if plot:
        plt.figure()
        plt.plot(xls, label = 'Function')
        plt.plot(xrs, label = 'Function')
        plt.grid(True)
        plt.xlabel('Iteration Number')
        plt.ylabel('Bounding x-Values')
    
    # return best estimate
    return 0.5 * (xll + xrr)
```

```python
# example usage
f = lambda x: x * (2 - x)
root = golden_section_search(f, 0.0, 2.0, true_err_max = 1e-6, debug = True, plot = True)
```
